```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
# Data Collection
stock_name = input("Enter the stock symbol (e.g., 'AAPL' for Apple, 'GOOGL' for Alphabet Inc.): ")
start_date = input("Enter the start date (YYYY-MM-DD): ")
end_date = input("Enter the end date (YYYY-MM-DD): ")
timeframe = input("Enter the timeframe (1d, 1wk, 1mo): ")

# Fetching data
stocks = yf.Ticker(stock_name)
data = stocks.history(start=start_date, end=end_date, interval=timeframe)
data.dropna(inplace=True)
```

```python
#SMA
data['SMA_5'] = data['Close'].rolling(window=5).mean()
plt.plot( data['SMA_5'], alpha=0.5)
plt.xlabel('Date')
plt.ylabel('SMA_5_Price')
plt.title(stock_name)
plt.show()
#EMA
data['EMA_5'] = data['Close'].ewm(span=5, adjust=False).mean()
plt.plot( data['EMA_5'], alpha=0.5)
plt.xlabel('Date')
plt.ylabel('EMA_5_Price')
plt.title(stock_name)
plt.show()
#MACD
data['EMA_12'] = data['Close'].ewm(span=12, adjust=False).mean()
data['EMA_26'] = data['Close'].ewm(span=26, adjust=False).mean()
data['MACD'] = data['EMA_12'] - data['EMA_26']
data['Signal_Line'] = data['MACD'].ewm(span=9, adjust=False).mean()
plt.plot(data.index, data['MACD'], label='MACD', color='blue')
plt.plot(data.index, data['Signal_Line'], label='Signal Line', color='red', linestyle='--')
plt.bar(data.index, data['MACD'] - data['Signal_Line'], label='MACD Histogram', color='gray',alpha=1)
plt.xlabel('Date')
plt.ylabel('MACD Value')
plt.title(stock_name)
plt.legend()
plt.grid(True)
plt.show()
data.dropna(inplace=True)
```

```python
# Scaling features and target
scaler_features = MinMaxScaler()
scaler_target = MinMaxScaler()
data['Target'] = data['Close']
scaled_features = scaler_features.fit_transform(data[['MACD', 'SMA_5', 'EMA_5']])
scaled_target = scaler_target.fit_transform(data[['Target']])
```

```python
#Sequence Creation
def sequence(data_features, data_target, sequence_length, prediction_length):
    X, y = [], []
    for i in range(sequence_length, len(data_features) - prediction_length + 1):
        X.append(data_features[i - sequence_length:i])
        y.append(data_target[i:i + prediction_length])
    return np.array(X), np.array(y)

sequence_len = 60
prediction_len = 15
X, y = sequence(scaled_features, scaled_target, sequence_len, prediction_len)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
#LSTM Model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(prediction_len))
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the model
history = model.fit(X_train, y_train,epochs=10,batch_size=32,validation_data=(X_test, y_test),verbose=1)
```

```python
# Step 5: Make Predictions
predicted_stock_price = model.predict(X_test)
y_test_actual = scaler_target.inverse_transform(y_test.reshape(-1, 1)).reshape(y_test.shape)
predicted_stock_price = scaler_target.inverse_transform(predicted_stock_price.reshape(-1, 1)).reshape(predicted_stock_price.shape[0], -1)
y_test_actual = np.nan_to_num(y_test_actual)
predicted_stock_price = np.nan_to_num(predicted_stock_price)
```

```python
# Plot predictions
plt.plot(y_test_actual.flatten(), label='Actual')
plt.plot(predicted_stock_price.flatten(), label='Predicted')
plt.title(f"{stock_name} Stock Price Prediction")
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
# Calculate R^2 Score
r2 = r2_score(y_test_actual.flatten(), predicted_stock_price.flatten())
print("R^2 Score:", r2)
```
